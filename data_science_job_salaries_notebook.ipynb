```python

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

## Load Dataset

```python

df = pd.read_csv('../dataset/Data Science Job Salaries.csv')
df.head()
```

## Dataset Overview

```python

print("Shape:", df.shape)
df.info()
```

```python

df.describe(include='all')
```

## Data Cleaning

```python

print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("\nNew Shape:", df.shape)
```

## Exploratory Data Analysis

### Salary Distribution

```python

sns.histplot(df['salary_in_usd'], bins=30, kde=True)
plt.title('Salary Distribution')
plt.show()
```

### Top Paying Jobs

```python

top_jobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Top Paying Jobs')
plt.show()
```

### Salary by Experience Level

```python

sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary by Experience Level')
plt.show()
```

### Salary Trend by Year

```python

salary_year = df.groupby('work_year')['salary_in_usd'].mean()
sns.lineplot(x=salary_year.index, y=salary_year.values, marker='o')
plt.title('Salary Trend by Year')
plt.show()
```

### Company Size Distribution

```python

company_size = df['company_size'].value_counts()
plt.pie(company_size.values, labels=company_size.index, autopct='%1.1f%%')
plt.title('Company Size Distribution')
plt.show()
```

### Remote Work Distribution

```python

sns.countplot(x='remote_ratio', data=df)
plt.title('Remote Work Distribution')
plt.show()
```

### Employment Type Analysis

```python

employment_salary = df.groupby('employment_type')['salary_in_usd'].mean()
employment_salary.plot(kind='bar')
plt.title('Average Salary by Employment Type')
plt.show()
```

### Top Company Locations

```python

top_locations = df['company_location'].value_counts().head(10)
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.xticks(rotation=45)
plt.title('Top Company Locations')
plt.show()
```

### Correlation Heatmap

```python

numeric_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Average Salary by Company Size

```python

company_salary = df.groupby('company_size')['salary_in_usd'].mean()
company_salary.plot(kind='bar')
plt.title('Average Salary by Company Size')
plt.show()
```


# Key Insights

- Executive-level professionals earn the highest salaries.
- Large companies generally offer better compensation.
- Remote work opportunities increased significantly after 2021.
- Data Analytics Lead and Principal Data Engineer roles are among the highest paid.
- The United States dominates the dataset in terms of company locations.
